## 1. Introduction

### A. Linear hypothesis function

```python
# numpy function performing dot product
def h(x, theta):
    y = np.dot(x, theta)
    return y
```

```python
# cheking the functionality of h(x, theta)
x = np.linspace(1,10,10) # [1.,2.,3.,...,10.]
t = np.array([1]*10)     # [1, 1, 1, ..., 1]
print(f"h(x,theta) = {h(x, t)}")
```

### B. Mean Squared Error fucntion

```python
def mean_squared_error(y_predicted, y_label):
    MSE = (1/len(y_label)) * np.sum((y_predicted - y_label)**2)
    return MSE
```

## 3. Closed-Form Solution

### A. LeastSquareRegression

```python
class LeastSquaresRegression():
    def __init__(self,):
        self.theta_ = None
        
    def fit(self, X, y):
        # Calculates theta that minimizes the MSE and updates self.theta_
        theta = np.linalg.multi_dot([np.linalg.inv(np.dot(np.transpose(X),X)), np.transpose(X), y])
        self.theta_ = theta
        
    def predict(self, X):
        # Make predictions for data X, i.e output y = h(X)
        y_pred = h(X, self.theta_)
        return y_pred
```

### B. Plotting the points

```python
# data
X = 4 * np.random.rand(100, 1)
y = 10 + 2 * X + np.random.randn(100, 1)
# plotting the data
plt.scatter(X, y)
plt.ylabel("y", fontsize=14)
plt.xlabel("x", fontsize=14)
```

### C. Addition of bias column

```python
def bias_column(X):
    X_new = np.append(np.ones((X.shape[0],1)), X, axis=1)
    return X_new
```

```python
# checking the functionality of bias_column(X)
X_new = bias_column(X)
print(X[:5])
print(" ---- ")
print(X_new[:5])
```

### D. Calculation of the weights

```python
# setting the model
model =LeastSquaresRegression()
# fitting the model
model.fit(X_new, y)
# printing the theta
print(model.theta_)
```

<font size="3">As expected, the coefficients are close to 10 and 2 for bias and theta, respectively.</font>

### E. Performing the predictions and plot comparison

```python
# predicting y values
y_new = model.predict(X_new)
```

```python
def my_plot(X, y, y_pred):
    plt.scatter(X, y, c='b', s=9, label="actual")
    plt.scatter(X, y_pred, c='r', s=9, label="predicted")
    plt.xlabel('x', fontsize=14)
    plt.ylabel('y', fontsize=14)
    plt.legend(fontsize=12)
    plt.show()
```

```python
# plotting actual y and predicted y
my_plot(X, y, y_new)
```

## 2. Gradient Descent

### A. GradientDescentOptimizer

```python
class GradientDescentOptimizer():
    def __init__(self, f, fprime, start, learning_rate = 0.1):
        self.f_ = f # function to be optimized
        self.fprime_ = fprime # derivitive of f
        self.current = start # current point to be evaluated
        self.learning_rate = learning_rate # learning rate
        # save history as attributes
        self.history = [start]
    
    def step(self):
        # Take a gradient descent step
        # 1. Compute the new value and update selt.current_
        current = self.current - self.learning_rate * self.fprime_(self.current)
        self.current = current
        # 2. Append the new value to history
        self.history.append(current)
        
    def optimize(self, iterations = 100):
        # Use the gradient descent to get closer to the minimum:
        # For each iteration, take a gradient step
        for i in range(iterations):
            self.step()
        
    def getCurrentValue(self):
        # Getter for current_
        return self.current
    
    def print_result(self):
        print("Best theta found is " + str(self.current))
        print("Value of f at this theta: f(theta) = " + str(self.f_(self.current)))
        print("Value of f prime at this theta: f'(theta) = " + str(self.fprime_(self.current)))
```

### B. Functions <i>f</i> and <i>f_prime</i> and functionality test

```python
# checking the functionality of the class GradientDescentOptimizer
def f(x):
    y = 3 + np.dot((x - np.array([2,6])), (x - np.array([2,6])).T)
    return y

def fprime(x):
    y_prime = 2 * (x - np.array([2,6]))
    return y_prime
```

```python
# optimization
grad = GradientDescentOptimizer(f, fprime, np.random.normal(size=(2,)), 0.1)
grad.optimize(10)
grad.print_result()
```

<font size="3">The Gradient Descent Optimization result demonstrate that x=[2,6] is the critical point of the function <i>f</i>. It is, indeed, a minimum point of the function <i>f</i>, since it is a quadratic function. The maximum value of f(x) is equal to 3, while the obtained derivitive value for the point is close to zero.<font>

### C. Plotting the functions and gradients

```python
# generating X map and calculating corresponding function f
X = np.mgrid[-4:8:30j, 0:12:30j].reshape(2,-1).T
Z = f(X)
x1, x2 = np.mgrid[-4:8:30j, 0:12:30j]

# plotting the function f as a 3d surface plot
fig = plt.figure(figsize=(10,10), dpi=80)
ax = plt.axes(projection='3d')
ax.contour3D(x1, x2, Z.diagonal().reshape(30,-1), 100, cmap='viridis', alpha=0.9)
ax.view_init(50, 30)
ax.set_xlabel('$x_1$',fontsize=12)
ax.set_ylabel('$x_2$',fontsize=12)
ax.set_zlabel('f(x)',fontsize=12)

# plotting the line that corresponds to each optimization step
X_history = np.array(grad.history)
Z_track = []
for x in X_history:
    Z_track.append(f(x))

ax.plot(X_history[:,0],X_history[:,1], Z_track, marker='o', c='r', markersize=4)
```

<font size="3"><b>How does the learning rate and the number of iterations influence the result ?</b><br>
    1) The learning rate affects the rate at which the model converges. Larger learning rate makes the optimizer converge faster; however, the optimizer wanders around minima not exactly reaching it. Thus, the final accuracy will suffer. Additionally, large learning rate may lead to divergence of the algorithm. On the other hand, smaller learning rate makes algorithm converge closer to minima; however, reaching the minima takes more updates as the steps are decreased.<br>
    2) More number of iterations generally leads to higher accuracy depending on selected learning rate. More iterations are better unless the model overfits.</font>

<font size="3"><b>How to tune hyperparameters and choose a good learning rate?</b><br>
Tuning of hyperparameters requires division of the dataset into train and validation sets first. Otherwise, the model will be overfitted to training dataset due to tuning. Generally, learning rate of 0.01-0.001 is selected as a thumb rule. However, depending on convergence and speed it can be altered. Commonly, dynamic learning rate are used, where learning rate is set to first value (say $l_r$ = 0.01) during initial <i>N</i> epochs, then it is reduced with an order of magnitude (say $l_r$ = 0.001) and trained for remaining of epochs for fine tuning.

</font>

<font size="3"><b>What about the number of iterations? What is the Convergence Rate?</b><br>
Depending on the value of the learning rate, different amount of iterations are needed to reach convergence. Generally, the number of training iterations is selected by observing the model performance over validation set. It helps to avoid overfitting, since a huge number of training iterations reduces the ability of the model to generalize, and consequently lead to poor performance in test set.
